```python
#calculate the number of rows and columns for subplot
num_plots=len(cat_cols)
num_rows=(num_plots+1)//2 #Add 1 and divide by 2 to round up for odd numbers 
num_cols=2

#Create a new figure 
mlt.figure(figsize=(20,25))  

#Loop through each feature and create a countplot 
for i, feature in enumerate(cat_cols,1):
    mlt.subplot(num_rows,num_cols,i)
    sns.countplot(x=feature,data=df,palette='dark')
    mlt.title(f'barplot for{feature}')
    mlt.xlabel(feature)
    mlt.ylabel('count')
    mlt.xticks(rotation=45)
#Adjust layout to prevent overlap of subplots 
mlt.tight_layout()
mlt.show()
```

```python
high_corr_cols=['emp.var.rate','euribor3m','nr.employed']
```

```python
df1=df.copy()
df1.columns
```

```python
df1.drop(high_corr_cols,inplace=True,axis=1)
df1.columns
```

```python
df1.shape
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df_encoded=df1.apply(lb.fit_transform)
df_encoded
```

```python
df_encoded['Recure'].value_counts()
```

```python
x=df_encoded.drop('Recure',axis=1)
y=df_encoded['Recure']
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x.test,y_train,y.test=train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x.test.shape)
print(y_train.shape)
print(y.test.shape)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

def eval_model(y_test,y_pred):
    acc=accuracy_score(y_test,y_pred)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))
def mscore (model):
    train_score=model.score(x_train,y_train)
    test_score=model.score(x.test,y.test)
    print('Training Score',train_score)
    print('Testing Score',test_score)



```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
mscore(dt)
```

```python
ypred_dt=dt.predict(x.test)
print(ypred_dt)
```

```python
eval_model(y.test,ypred_dt)
```
